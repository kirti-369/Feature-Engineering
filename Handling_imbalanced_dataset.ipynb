Handling Imbalanced Dataset


1.   Upsampling
2.   Down sampling




```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(123)
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0
```

```python
n_class_1
```

```python
n_class_0,n_class_1
```

```python
class_0 = pd.DataFrame({
    'feature1':np.random.normal(loc=0,scale=1,size=n_class_0),
    'feature2':np.random.normal(loc=0,scale=1,size=n_class_0),
    'target':[0]*n_class_0
})

class_1 = pd.DataFrame({
    'feature1': np.random.normal(loc=2,scale=1,size=n_class_1),
    'feature2': np.random.normal(loc=2,scale=1,size=n_class_1),
    'target':[1]*n_class_1
})
```

```python
df = pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df['target'].value_counts()
```

```python
##Upsampling
```

```python
df_minority = df[df['target']==1]
df_majority = df[df['target']==0]
```

```python
df_minority.head()
```

```python
df_majority.head()
```

```python
##For performing upsampling import sklearn
```

```python
from sklearn.utils import resample
```

```python
df_minority_upsampled = resample(df_minority,
                                 replace=True,
                                 n_samples=len(df_majority),#to match the majority class
                                 random_state=42)
```

```python
df_minority_upsampled.shape
```

```python
df_minority_upsampled['target'].value_counts()
```

```python
df_unsampled = pd.concat([df_majority,df_minority_upsampled])
```

```python
df_unsampled.head()
```

```python
df_unsampled.shape
```

```python
##Downsampling
```

```python
class_0 = pd.DataFrame({
    'feature1':np.random.normal(loc=0,scale=1,size=n_class_0),
    'feature2':np.random.normal(loc=0,scale=1,size=n_class_0),
    'target':[0]*n_class_0
})

class_1 = pd.DataFrame({
    'feature1': np.random.normal(loc=2,scale=1,size=n_class_1),
    'feature2': np.random.normal(loc=2,scale=1,size=n_class_1),
    'target':[1]*n_class_1
})
```

```python
pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df_minority = df[df['target']==1]
df_majority = df[df['target']==0]
```

```python
df_majority_downsample = resample(df_majority,
                                 replace=False,
                                 n_samples=len(df_minority),#to match the majority class
                                 random_state=42)
```

```python
df_majority_downsample
```

```python
df_majority_downsample.shape
```

```python
df_downsample = pd.concat([df_minority,df_majority_downsample])
```
